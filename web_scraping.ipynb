```python
import requests
import pandas as pd
import numpy as np
```

### 1. Download FII's with requests

```python
url = 'https://www.fundsexplorer.com.br/ranking'
headers = {
    'User-Agent': 
        'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_4) AppleWebKit/537.36'
        ' (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'
}
```

```python
response = requests.get(url, headers=headers)
```

```python
if response.status_code == 200:
    df_fiis = pd.read_html(response.content, encoding='utf-8')[0]
```

```python
df_fiis.info()
```

```python
df_fiis.describe(include='all')
```

### 2. Cleaning and formatting

```python
df_fiis['Setor'].unique()
```

```python
df_fiis = df_fiis.replace('Títulos e Valores Mobiliários', 'Títulos e Val. Mob.', regex=False)
df_fiis['Setor'].unique()
```

```python
df_fiis.columns
```

```python
df_fiis.isna().sum()
```

```python
categorical_columns = ['Código do fundo','Setor']
idx = df_fiis[df_fiis['Setor'].isna()].index
df_fiis.drop(idx, inplace=True)
df_fiis[categorical_columns].isna().sum()
```

```python
df_fiis[categorical_columns] = df_fiis[categorical_columns].astype('category')
df_fiis.info()
```

```python
df_fiis.isna().sum()
```

```python
col_floats = list(df_fiis.iloc[:,2:-1].columns)
df_fiis[col_floats] = df_fiis[col_floats].fillna(value=0)
df_fiis[col_floats]
```

```python
df_fiis[col_floats] = df_fiis[col_floats].applymap(lambda x: str(x).replace('R$', '').replace('.0','').replace('.','').replace('%','').replace(',','.'))
df_fiis[col_floats] = df_fiis[col_floats].astype('float')
df_fiis['Liquidez Diária'] = df_fiis['Liquidez Diária'].astype('int64')
df_fiis.info()
```

```python
df_fiis.describe()
```

```python
df_fiis['Vacância Física'] = df_fiis['Vacância Física']/100
df_fiis['Vacância Financeira'] = df_fiis['Vacância Financeira']/100
df_fiis['P/VPA'] = df_fiis['P/VPA']/100

tipo = []
for row in df_fiis['Setor']:
    if row in ('Shoppings','Lajes Corporativas','Residencial','Logística','Hospital','Hotel'):
        tipo.append('Tijolo')
    else: tipo.append('Papel')

df_fiis.insert(1,'Tipo',tipo)

df_fiis.head()
```

```python
df_fiis.to_csv("fiis.csv", index=False)
```
