```python
import pandas as pd
```

```python
df_fiis = pd.read_csv('fiis.csv')
```

```python
df_fiis.columns
```

```python
indicators = ['Código do fundo',
               'Tipo',
               'Setor',
               'Preço Atual',
               'Dividend Yield',
               'DY (12M) Acumulado', 
               'Vacância Física', 
               'Vacância Financeira', 
               'P/VPA', 
               'Quantidade Ativos', 
               'Liquidez Diária']
df_fiis = df_fiis[indicators]
df_fiis
```

```python
def best_papers(df_fiis):

    filters = (df_fiis['Tipo'] == 'Papel') &\
        (df_fiis['Dividend Yield'] >= 1) &\
        (df_fiis['DY (12M) Acumulado'] >= 12) &\
        (df_fiis['P/VPA'] < 1.0) &\
        (df_fiis['Liquidez Diária'] > 5000)

    return df_fiis[filters].nlargest(10,['Dividend Yield','DY (12M) Acumulado'])
```

```python
def best_bricks(df_fiis):

    filters = (df_fiis['Tipo'] == 'Tijolo') &\
        (df_fiis['Vacância Física'] <= 0.05) &\
        (df_fiis['Vacância Financeira'] <= 0.05) &\
        (df_fiis['Liquidez Diária'] >= 1000) &\
        (df_fiis['Quantidade Ativos'] >= 5)
    
    return df_fiis[filters].nlargest(10,['Dividend Yield','DY (12M) Acumulado'])
```

```python
df_best_papers = best_papers(df_fiis)
df_best_papers.reset_index(drop=True, inplace=True)

df_best_bricks = best_bricks(df_fiis)
df_best_bricks.reset_index(drop=True, inplace=True)
```
